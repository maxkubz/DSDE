```python
import os
import json
directory = ['./drive/MyDrive/Project/2018/',
 './drive/MyDrive/Project/2019/',
 './drive/MyDrive/Project/2020/',
 './drive/MyDrive/Project/2021/',
 './drive/MyDrive/Project/2022/',
 './drive/MyDrive/Project/2023/']

for dir in directory:
  files = os.listdir(dir)
  for f in files:
    pathName = dir + f
    with open(pathName,'r') as data:
      json_file = json.load(data)
      authors = json_file.get("abstracts-retrieval-response" , {}).get('item',{}).get("bibrecord",{}).get('head',{}).get("author-group",{})
      if isinstance(authors, list):
        for author in authors:
          if isinstance(author.get("affiliation"), str):
            continue
          affiliation = author.get("affiliation")
          if affiliation:  # Check if affiliation is not None
            countryName = affiliation.get("country")
            cityName = affiliation.get("city")
            organizations = affiliation.get("organization", [])
            if organizations:
              for e in organizations:
                if isinstance(e, dict) and "$" in e:
                  organization = e["$"]
                  allGivenData.append([organization, cityName, countryName])
                  print([organization, cityName, countryName] , "Added!")
      else:
        affiliation = authors.get("affiliation")
        if(affiliation):
          countryName = affiliation.get("country") if "country" in affiliation else None
          cityName = affiliation.get("city") if "city" in affiliation else None
          organizations = affiliation.get("organization", [])
          if organizations:
            for e in organizations:
              if isinstance(e, dict) and "$" in e:
                organization = e["$"]
                allGivenData.append([organization, cityName, countryName])
                print([organization, cityName, countryName] , "Added!")
```

```python
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder \
    .appName("Read JSON File") \
    .getOrCreate()

# Load JSON file into a DataFrame
df = spark.read.csv("./drive/MyDrive/result_csv.csv" , header = True)
df2 = spark.read.csv("./drive/MyDrive/givenData.csv" , header = True)
# Show the DataFrame schema
df.printSchema()
df2.printSchema()

# Show the first few rows of the DataFrame
df.show()
df2.show()

# Perform inner join based on a common column
merged_df = df.union(df2)

# Show the merged DataFrame
print("Merged DataFrame:")
merged_df.show()

print(df.count())
print(df2.count())
print(merged_df.count())
```

```python
merged_df = merged_df.dropna(how='any', subset=None) # Drop if any value is missing
```
